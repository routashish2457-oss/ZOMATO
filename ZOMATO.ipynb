# INPUT AND OUTPUT

```python
zomato.columns
```

```python
ip = zomato.drop(['book_table','name', 'rate'], axis=1)
op = zomato.rate
```

```python
ip.head()
```

```python
op.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,train_size=0.8)
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

# CONFUSION MATRIX

```python
#Accuracy
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
acc = accuracy_score(pred, y_test)
recall = recall_score(pred, y_test, average='weighted')
precision = precision_score(pred, y_test, average='weighted')
f1 = f1_score(pred, y_test, average='weighted')
print('Accuracy:  ', acc)
print('Recall:  ', recall)
print('Precision:  ', precision)
print('F1_Score:  ', f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(pred,y_test)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```

# Naive-Bayes

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(x_train,y_train)
```

```python
pred1 = gn.predict(x_test)
pred1
```

```python
#Accuracy
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
acc = accuracy_score(pred1, y_test)
recall = recall_score(pred1, y_test, average='weighted')
precision = precision_score(pred1, y_test, average='weighted')
f1 = f1_score(pred1, y_test, average='weighted')
print('Accuracy:  ', acc)
print('Recall:  ', recall)
print('Precision:  ', precision)
print('F1_Score:  ', f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm1 = confusion_matrix(pred1,y_test)
cm1
```

```python
cmd = ConfusionMatrixDisplay(cm1)
cmd.plot()
plt.show()
```
